Navigate the grid to find out the numbers and symbols. Then use [breadth first search](https://en.wikipedia.org/wiki/Breadth-first_search) to find the shortest path.

````python
TARGET = 30
graph = [*map(str.split, [
    "  *  8   -   1 ",
    "  4  *  11   * ",
    "  +  4   -  18 ",
    " 22  -   9   * ",
])]

queue = [(int(graph[3][0]), 3, 0, "")]
visited = set()
while queue:
    n, y, x, path = queue.pop(0)
    
    if (n, y, x) in visited:
        continue
    visited.add((n, y, x))
    
    if (y, x) == (0, 3):
        if n == TARGET:
            print(path)
            break
        continue
    
    for r, (ay, ax) in zip("nswe", [(y-1,x),(y+1,x),(y,x-1),(y,x+1)]):
        if ay not in range(4) or ax not in range(4):
            continue
        for s, (by, bx) in zip("nswe", [(ay-1,ax),(ay+1,ax),(ay,ax-1),(ay,ax+1)]):
            if by not in range(4) or bx not in range(4) or (by, bx) == (3, 0):
                continue
            
            v = eval(str(n) + graph[ay][ax] + graph[by][bx])
            queue.append((v, by, bx, path + r+s))
````
